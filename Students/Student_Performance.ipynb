```python
# ading new column to dataset
df['Total score'] = df['math score'] + df['reading score'] + df['writing score']
df['Average score'] = df['Total score'] / 3
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
plt.subplots(1,2,figsize =(15,7))# 1,2 gives 1 row 2 columns and figsize is figure size
plt.subplot(1,2,1)# 1,,2,1 means 1,2 ko 1 vaney ko (first)
sns.histplot(data = df,x= 'Average score', bins =40 , kde = True, color ='g')# color g means green
plt.subplot(1,2,2)
sns.histplot(data = df, x='Average score', bins = 40, kde = True, hue = 'gender')
plt.show()
```

```python
plt.subplots(1,3,  figsize= (25,6))
plt.subplot(1,3,1)
sns.histplot(data = df, x= 'Average score',kde = True, hue = 'lunch')
plt.subplot(1,3,2)
sns.histplot(data = df[df['gender']=='female'], x = 'Average score', kde = True, hue = 'lunch')
plt.subplot(1,3,3)
sns.histplot(data = df[df['gender']=='male'], x= 'Average score',kde = True, hue = 'lunch')
plt.show()
```

### Insights:
     Standard lunch helps students perform well in exam
     Standard lunch helps perform well in exams be it male or female.

```python
plt.subplots(1,3, figsize = (25,6))
plt.subplot(1,3,1)
sns.histplot(data = df, x = 'Average score',kde = True, hue = 'race/ethnicity')
plt.subplot(1,3,2)
sns.histplot(data = df[df['gender']=='female'], x= 'Average score', kde = 'True', hue = 'race/ethnicity')
plt.subplot(1,3,3)
sns.histplot(data = df[df['gender']=='male'], x= 'Average score', kde = 'True', hue = 'race/ethnicity')
plt.show()
```
